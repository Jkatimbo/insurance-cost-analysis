#  Linear Regression Modeling

Now we will fit in a linear regression model to the data and visualize the best_fit line:

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Assuming 'insurance_df' is your DataFrame
# You should replace this with the actual DataFrame you are working with
insurance_df = pd.read_csv("/Users/Admin/Desktop/insurance.csv")  # Replace with the actual path

# Check if 'age' column exists in the DataFrame
if 'age' in insurance_df.columns and 'charges' in insurance_df.columns:
    # 'age' and 'charges' columns exist
    features = insurance_df[['age']]  # Selecting 'age' as a feature
    target = insurance_df['charges']  # Target variable 'charges'
    
    # Create and fit the Linear Regression model
    model = LinearRegression()
    model.fit(features, target)
    
    # Make predictions
    predictions = model.predict(features)
    
    # Plot the scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(insurance_df['age'], insurance_df['charges'], color='blue', alpha=0.7, label='Scatter Plot')
    
    # Plot the best fit line
    plt.plot(insurance_df['age'], predictions, color='red', linewidth=2, label='Best Fit Line')
    
    plt.title('Scatter Plot of age vs. Charges with Best Fit Line')
    plt.xlabel('age')
    plt.ylabel('charges')
    plt.legend()
    plt.show()
else:
    # 'age' or 'charges' column does not exist
    print("Columns 'age' or 'charges' do not exist in the DataFrame.")
    # You may want to handle this case or check for other columns
```
